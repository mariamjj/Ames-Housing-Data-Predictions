# Ames Housing Data Predictions

In this project, we will evaluate the performance and predictive power of a model that has been trained and tested on data collected from homes in suburbs of **Ames, Iowa**. A model trained on this data that is seen as a *good fit* could then be used to make certain predictions about a home — in particular, its monetary value. This model would prove to be invaluable for someone like a real estate agent who could make use of such information on a daily basis.

The Ames housing data has 879 entries that represent aggregated data for 80 features for homes from various suburbs in Ames, Iowa from 2006 to 2010. 

## Problem Statement

We are interested in evaluating the price of a house on sale in Ames, IA by using a regression model on the Ames Housing Dataset and: i) predicting what the sale price will be and ii) determining what variables (if any) effect the final sale price.

## Model on Training and Validation Test Set

```python
features = ['TotalLivingSF * Overall Qual * Year Built', 'squared', 'TotalLivingSF * Overall Qual', 'TotSFCars','TotalLivingSF * Year Built', 'Overall1st' ,'Overall Qual^2',
              'TotalLivingSF','TotSFnGarage', 'Gr Liv Area * Total Bsmt SF', 'Overall Qual','TotLivTotBasement', 'Gr Liv Area']
```

```python
X = train[features]
y = train['SalePrice']
```

```python
def prep_model(df,X,y,test_split_num):
    ss = StandardScaler()
    
    #train/test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= test_split_num, random_state = 42)
    
    #standard scaling
    X_train = ss.fit_transform(X_train)
    
    X_test = ss.transform(X_test)
    
    return X_train, X_test, y_train, y_test
```

```python
def train_LR_model(df, X_train, y_train):
    lr = LinearRegression()
    
    #fit and score training set
    model = lr.fit(X_train, y_train)
    train_score = lr.score(X_train, y_train)
    coef = lr.coef_
    
    #pickle model
    dump(model, 'final.joblib')
    
    print('The Lin Reg model was fit to the training set and returned a R^2 score of: {}'.format(train_score))
    
    return model, train_score, coef
```

```python
#Create a function for ridge
def train_ridge_model(df, X_train, y_train):
    alpha_list = np.logspace(0,5,200)
    ridge_model = RidgeCV(alphas = alpha_list, store_cv_values = True)
    
    #fit model
    ridge_model = ridge_model.fit(X_train, y_train)
    
    #highest alpha for best result
    best_alpha = ridge_model.alpha_
    ridge_opt_model = Ridge(alpha = best_alpha)
    
    #r score
    model2 = ridge_opt_model.fit(X_train,y_train) #why do we have to fit this again??
    r2_ridge = model2.score(X_train, y_train)
    
    print('The Ridge model was fit to the training set and returned a R^2 score of: {}'.format(r2_ridge))
    
    return model2, r2_ridge, best_alpha
```

```python
#Create a function for lasso
def train_lasso_model(df, X_train, y_train):
    l_alpha_list = np.arange(0,5,0.400)
    lasso_model = LassoCV(alphas= l_alpha_list, cv = 5)
    
    #fit model
    lasso_model = lasso_model.fit(X_train, y_train)
    
    #highest alpha
    best_alpha = lasso_model.alpha_
    lasso_opt_model = Lasso(alpha = best_alpha)
    
    #r score
    model2 = lasso_opt_model.fit(X_train,y_train)
    r2_lasso = model2.score(X_train, y_train)
    
    print('The Lasso model was fit to the training set and returned a R^2 score of: {}'.format(r2_lasso))
    
    return model2, r2_lasso, best_alpha
```

```python
#Create a function for elastic net
def train_enet_model(df, X_train, y_train):
    enet_alphas = np.arange(0.5, 1.0, 0.005)
    enet_model = ElasticNetCV(alphas = enet_alphas)
    
    #fit model
    enet_model = enet_model.fit(X_train, y_train)
    
    #highest alpha
    best_alpha = enet_model.alpha_
    enet_opt_model = ElasticNet(alpha=best_alpha)
    
    #r score
    model2 = enet_opt_model.fit(X_train, y_train)
    r2_enet = model2.score(X_train, y_train)
    
    print('The Elastic Net model was fit to the training set and returned a R^2 score of: {}'.format(r2_enet))
    
    return model2, r2_enet, best_alpha
```

```python
def knn_reg_model(df, X_train, y_train):
    knn = KNeighborsRegressor()
    
    #fit model
    model2 = knn.fit(X_train, y_train)
    
    #score
    r2_knn = knn.score(X_train, y_train)
    
    print('The KNN Regressor was fit to the training set and returned a R^2 score of: {}'.format(r2_knn))
    
    return model2, r2_knn
```

```python
#Create a function for cross validation and grid search
def gridsearch(model_name, X_train, y_train, cv_numb):
    #cv_accuracies = cross_val_score(model, X_train, y_train, cv_numb).mean()
    
    if model_name == 'knn_model':
        
        knn_params = {'n_neighbors': range(1, 101, 10),'p': [1, 2],'weights': ['distance', 'uniform']}
        
        knn_gridsearch = GridSearchCV(KNeighborsClassifier(), knn_params, cv = cv_numb, verbose = 0)
        
        grid_model = knn_gridsearch.fit(X_train, y_train)
        knn_gridsearch.best_score_
        knn_gridsearch.best_params_
        best = knn_gridsearch.best_estimator_
        score = best.score(X_train, y_train)
        
        print('After tuning the hyper parameters, the best KNN score was: {}'.format(score))
        
    elif model_name == 'ridge_model':
        
        ridge_params = {'alphas': np.logspace(0,2,400), 'store_cv_values': True}
        
        ridge_gridsearch = GridSearchCV(RidgeCV, ridge_params, cv = cv_numb, verbose = 0)
        
        grid_model = ridge_gridsearch.fit(X_train, y_train)
        ridge_gridsearch.best_score_
        ridge_gridsearch.best_params_
        best = ridge_gridsearch.best_estimator_
        score = best.score(X_train, y_train)
        
        print('After tuning the hyper parameters, the best ridge score was: {}'.format(score))
        
    elif model_name == 'lasso_model':
        lasso_params = {'alphas': np.arange(200,400,0.5), 'cv':[5,10,15] }
        
        lasso_gridsearch = GridSearchCV(LassoCV, lasso_params, cv = cv_numb, verbose = 0)
        
        grid_model = lasso_gridsearch.fit(X_train, y_train)
        lasso_gridsearch.best_score_
        lasso_gridsearch.best_params_
        best = lasso_gridsearch.best_estimator_
        score = best.score(X_train, y_train)
        
        print('After tuning the hyper parameters, the best lasso score was: {}'.format(score))
        
    else:
        enet_params = {'alphas': np.arange(0, 400, 0.005)}
        
        enet_gridsearch = GridSearchCV(ElasticNetCV, enet_params, cv = cv_numb, verbose = 0)
        
        grid_model = enet_gridsearch.fit(X_train, y_train)
        enet_gridsearch.best_score_
        enet_gridsearch.best_params_
        best = enet_gridsearch.best_estimator_
        score = best.score(X_train, y_train)
        
        print('After tuning the hyper parameters, the best enet score was: {}'.format(score))
        
    return grid_model
```

```python
def test_model(df, X_test, model):
    
    #calculate predictions and r2 score
    predictions = model.predict(X_test)
    
    return predictions
```

```python
def regression_metrics(df, model, X_test, y, predictions):
    r_squared = metrics.r2_score(y, predictions)
    mean_squared_err = metrics.mean_squared_error(y, predictions)
    root_mean_squared_err = np.sqrt(mean_squared_err)
    mean_abs_err = metrics.mean_absolute_error(y, predictions)
    median_abs_err = metrics.median_absolute_error(y, predictions)
    adj_r_squared = 1 - (1 - r_squared)*(len(y)-1)/(len(y)-X_test.shape[1]-1)
    
    print("The model returned a R^2 of: {} for the validation test".format(r_squared))
    print("Adjusted R-squared: {}".format(adj_r_squared))
    print("Mean Squared Error: {}".format(mean_squared_err))
    print("Root Mean Squared Error: {}".format(root_mean_squared_err))
    print("Mean Absolute Error: {}".format(mean_abs_err))
    print("Median Absolute Error: {}".format(median_abs_err))
    
    return r_squared
```

```python
def plot(df, y_test, features, coef, predictions):
    residuals = y_test - predictions
    #bar graph - show the correlation of the coef to see if it's positive or neg
    pd.Series(coef, index = features).plot.bar(figsize = (15,10))
    plt.show()
    
    #scatter
    plt.scatter(predictions, residuals)
    plt.axhline(0, color = 'red')
    plt.xlabel('Predictions')
    plt.ylabel('Residuals')
    plt.title('How much we missed by')
    plt.show()
    
    # true price vs predicted price
    plt.subplots(figsize=(15,7))
    plt.scatter(x=y_test, y=predictions, c='k')
    plt.plot(predictions, predictions, color='k') #draw line through center
    plt.xlabel('True price')
    plt.ylabel('Predicted price')
    plt.title('True price vs Predicted price for Validation Training Data')
    plt.show()
```
